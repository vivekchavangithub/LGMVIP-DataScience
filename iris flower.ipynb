# 6 - dropping the 'species' column

```python
A = df.drop(['species'],axis=1)
A
```

# 7 - Converting the 'species' column into a one-hot numeric column

```python
Label_Encode = LabelEncoder()
B = df['species']
B = Label_Encode.fit_transform(B)
```

```python
B
```

```python
df['species'].nunique()
```

```python
A = np.array(A)
A
```

# 8 - spliting the dataset into train & test data 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(A, B, test_size = 0.3, random_state=0)
```

```python
X_train
```

```python
X_test.shape
```

# 9 - Converting the train & test data into standard format

```python
from sklearn.preprocessing import StandardScaler
standard_scaler = StandardScaler().fit(X_train)
X_train_std = standard_scaler.transform(X_train)
X_test_std = standard_scaler.transform(X_test)
```

```python
X_train_std
```

# 10 - Model Preparation

## KNN Algorithm

```python
Y_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train_std,Y_train)
```

```python
predict_knn=knn.predict(X_test_std)
accuracy_knn=accuracy_score(Y_test,predict_knn)*100
```

```python
accuracy_knn
```
